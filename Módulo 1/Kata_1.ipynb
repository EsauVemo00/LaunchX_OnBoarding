## Convertidor de Unidades
Realizar un convertidor de unidades de Parsecs y Años luz

```python
Parsec=input('Cantidad de Parsecs a convertir')
Par2Ly=3.26156
lightyear=int(Parsec)*Par2Ly
print(str(Parsec) + ' Parsec, son '+ str(lightyear)+' lightyears')
```
