## Expresiones de prueba (ELSE)

```python
a=93
b=27
if a>=b:
    print(a)
else:
    print(b)
```

## Expresiones de prueba (ELIF)

```python
a=93
b=27
if a>=b:
    print('a es mayor o igual que b')
elif a==b:
    print('a es igual que b')
```

## Combinaciar declaraciones

```python
a=93
b=27
if a>b:
    print('a es mayor que b')
elif a<b:
    print('a es menor que b')
else:
    print('a es igual que b')
```

## Lógica condicional anidada

```python
a = 16
b = 25
c = 27
if a > b:
    if b > c:
        print ("a es mayor que b y b es mayor que c")
    else: 
        print ("a es mayor que b y menor que c")
elif a == b:
    print ("a es igual que b")
else:
    print ("a es menor que b")
```

# Operadores (AND y OR)
